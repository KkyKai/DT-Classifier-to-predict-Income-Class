### Implementing splitting function to split the dataset into a train and test set

```python
# Split a dataset into a train and test set
def train_test_split(dataset, test_size):
    if isinstance(test_size,float):
        #calculate the number of rows for test set based on test size
        test_size = round(test_size * len(dataset))
    
    #All rows in df to list
    indices = dataset.index.tolist()
    
    # A random sample of rows added to the test set based on test_size.
    test_indices = random.sample(population=indices, k=test_size)
    
    #update indices list
    indices = [ele for ele in indices if ele not in test_indices]
    
    #access rows with test_indices
    test = dataset.loc[test_indices]
    
    #Access rows without test_indices
    train = dataset.drop(test_indices)
 
    return train, test
```

## 1. Function for calculating information gain (entropy)

```python
# calculate entropy/info gain for the dataset

#Last column of dataframe is usually the label column
label_col  = train_df.iloc[:, -1]

#Find unique values and frequency in label column 
_, n_counts = np.unique(label_col, return_counts=True)

#Probability of unique values (frequency of value/sum of all frequencies)
prob = n_counts / n_counts.sum()

#Entropy = sum of (probability * -log2(probabilities))
entropy_of_dataset = sum(prob * -np.log2(prob))

print("Entropy of dataset: ", entropy_of_dataset)
```

#### Measuring impurity of dataset. An entropy of ~0.8 means that our dataset is relatively impure, and that our data is widely spread out among our classes.

```python
#Function for calculating entropy
def calculate_entropy(data):
    
    #Last column of dataframe is usually the label column
    label_col  = data[:, -1]
    
    #Find unique values and frequency in label column 
    _, n_counts = np.unique(label_col, return_counts=True)

    #Probability of unique values (frequency of value/sum of all frequencies)
    prob = n_counts / n_counts.sum()
    
    #Entropy = sum of (probability * -log2(probabilities))
    entropy = sum(prob * -np.log2(prob))
     
    return entropy
```

### Function for calculating entropy of splits

```python
def calculate_child_entropy(left, right):
    
    # calculate total number of rows
    n = len(left) + len(right)
    
    # calculate the occurences of each feature
    p_left = len(left) / n
    p_right = len(right) / n

    # calculate weighted entropy based on the probabilities of each feature
    child_entropy =  (p_left * calculate_entropy(left) + p_right * calculate_entropy(right))
    
    return child_entropy
```

## 2. Function for calculating gini impurity

### Function for calculating gini impurity for label

```python
#calculate gini for target
def calculate_gini(data):
    
    #All unique class labels
    class_labels = np.unique(data)
    
    gini = 0
    
    for classes in class_labels:
        
        #probability of a record belonging to each class 
        p_classes = len(data[data == classes]) / len(data)
        
        #Add to gini impurity
        gini += p_classes**2
        
    return 1 - gini
```

### Function for calculating gini impurity for feature

```python
def calculate_gini_gain(parent, left, right):
    
    #Weight of left node
    left_weight = len(left) / len(parent)
    
    #Weight of right node
    right_weight = len(right) / len(parent)
    
    #Gini gain = gini impurity of the parent node - the weighted sum of the gini impurities of the two child nodes
    giniGain = calculate_gini(parent) - (left_weight *calculate_gini(left) + right_weight *calculate_gini(right))
    
    return giniGain
```

#### The function calculates and returns the gini impurity gain, which is the improvement/reduction in homogeneity that results from splitting the "parent" data into two child nodes "left" and "right". 

## 3. Function for getting all possible splits

```python
def get_possible_splits(data):
    possible_splits = {}
    
    #Get no. of rows n columns using shape property
    n_rows , n_cols = data.shape
    
    #Get all possible splits for each feature except targe
    for col_index in range(n_cols-1):
        possible_splits[col_index] = {}
        values = data[:, col_index]
        unique_values = np.unique(values)
    
        possible_splits[col_index] = unique_values
                
    return possible_splits
```

#### The "possible_splits" dictionary returned by the function contains all the possible split points for each feature column in the dataframe, which can be used to find the best split that maximizes the gini impurity/info/gain ratio gain.

### Function for binary splits

```python
# split the dataframe into two child nodes produced by split
def split_data(data, split_col, split_val):
    
    split_col_val = data[:, split_col]
    
    left = data[split_col_val == split_val]
    right = data[split_col_val !=  split_val]
    
    return left, right
```

## Function for calculating best split based on information gain

```python
#Function for calculating best split based on information gain
def determine_best_split_infogain(data, possible_splits):
    
    # calculate the entropy of the data
    parent_entropy = calculate_entropy(data)
    best_entropy = parent_entropy

    # loop through all values for each column
    for col_index in possible_splits:
        
        for value in possible_splits[col_index]:
            
            #Split data into 2 nodes
            left, right = split_data(data, split_col=col_index, split_val=value)
            
            #Calculate entropy of split
            child_entropy = calculate_child_entropy(left,right)
            
            # compare each entropy value and find lowest
            if child_entropy < best_entropy:
                best_split_col = col_index
                best_split_val = value
                best_entropy = child_entropy
                info_gain = parent_entropy - best_entropy
    
    return best_split_col, best_split_val
```

```python
def determine_best_split_gini_index(data, possible_splits):
    
    label_col  = data[:, -1]
  
    # initialise max gini for comparision
    max_gini_gain = -float("inf")

    
    #dictionary to store possible splits and corresponding gini index
    gini = {}
    
    # loop through all values for each column
    for col_index in possible_splits:
        for value in possible_splits[col_index]:
            
            left, right = split_data(data, split_col=col_index, split_val=value)
            if len(left)>0 and len(right)>0:            
                y, left_y, right_y = data[:, -1], left[:, -1], right[:, -1]
                #calculate gini for each feature subset
                curr_gini_gain = calculate_gini_gain(y, left_y, right_y)

                ##feature with highest gini gain is selected as the splitting feature
                # compare each subset's value
                if curr_gini_gain > max_gini_gain:
                    max_gini_gain = curr_gini_gain
                    best_split_col = col_index
                    best_split_val = value

    return best_split_col, best_split_val
```

## Function for calculating best split based on gain ratio

```python
def determine_best_split_gain_ratio(data, possible_splits):
    #Calculate the entropy of the data
    parent_entropy = calculate_entropy(data)

    # Initialise best entropy and gain ratio
    best_entropy = parent_entropy
    best_gain_ratio = 0

    #Loop over each column and values in possible splits
    for col_index in possible_splits:
        
        for value in possible_splits[col_index]:
            #Split data into 2 nodes
            left, right = split_data(data, split_col=col_index, split_val=value)

            #Get child entropy for split
            child_entropy = calculate_child_entropy(left, right)

            #Get split info for the current split
            split_info = calculate_Split_info(possible_splits[col_index])

            #Compute info gain for split
            info_gain = parent_entropy - child_entropy

            # Compute the gain ratio for the split
            gain_ratio = info_gain / (split_info + 1e-10)

            #compare child entropy
            #if child_entropy < best_entropy AND
            #compare gain ratio
            #feature with highest gain ratio will be used to split
            if child_entropy < best_entropy and gain_ratio > best_gain_ratio:
                best_entropy = child_entropy
                best_gain_ratio = gain_ratio
                best_split_col = col_index
                best_split_val = value

    return best_split_col, best_split_val
```

```python
# Function to calculate split information for one feature
def calculate_Split_info(data_col):
    
    # Initialize the split info
    split_info = 0
    
    # Get the unique values and the count of each value in feature
    unique_vals, n_counts = np.unique(data_col, return_counts=True)
    
    # Probability of each unique value
    prob = n_counts / n_counts.sum()
    
    for idx, values in enumerate(unique_vals):

        # Calculate the number of occurrences of the unique value
        occurences = n_counts[idx]
        
        prob = occurences/n_counts.sum()
        temp = (prob* - np.log2(prob))
        split_info = split_info + temp

    return(split_info)
```

## Function for checking if we have reached a leaf node

```python
def check_leaf(data):
    #store last column as label_col
    label_col = data[:,-1]
    
    #extract unique values from label_col
    classes = np.unique(label_col)

    # If there is only one unique value, the data is a leaf node
    if len(classes) == 1:
        return True
    else:
        return False
```

## Function for classification

```python
#function gets the class for when a leaf node is reached
def classify_data(data):
    
    #since labels are always in the last column
    label_col = data[:,-1]
    
    #find each unique classes in the label column and how many times it appears
    classes, n_classes = np.unique(label_col, return_counts=True)

    #find index of the most frequent class
    index = n_classes.argmax()
    
    #extract out the name
    classification = classes[index]
    
    return classification
```

## Function for predicting values using the tree

```python
#Get label prediction for 1 row
def classify_one(test, tree):
    
    # save tree dictionary into list
    #question = keys
    question = list(tree.keys())[0]
    
    # extract out feature label, operator'=' and values of the feature
    # feature_name = key, operator = "=", value = value of key
    feature_name, operator, value = question.split(" ")

    # asking the question
    #if the sample value matches the question's value, it moves on to the left subtree, otherwise it moves to the right subtree. 
    if str(test[feature_name]) == value:
        answer = tree[question][0]
    else:
        answer = tree[question][1]

   
    #if it is a dictionary, it means that the answer is a subtree.
    #recursively call method until answer is no longer a dictionary, and hence found
    #else, it is the answer
    #then return answer as final prediction
    
    #If not dictionary return ans as prediction
    if not isinstance(answer, dict):
        return answer
    
    #Else recursively call function until ans is found
    else:
        residual_tree = answer
        return classify_one(test, residual_tree)
```

```python
#Get a list of predictions for all rows in test_df
def predict(test,tree):
    #initilaise array for storing 
    output = []
    index = test.index.tolist()
    
    #Loop through each row of test_df and apply classify_one function to predict label for row
    for i in index:
        output.append(classify_one(test.loc[i],tree))
        
    #Returns a list of predictions for each row in test_df.
    return output
```

## Function for calculating prediction accuracy

```python
#By comparing all predictions to its actual values, calculating the average mean of total predictions
def calculate_accuracy(df, tree):

    #Add a new column to df with the result of the prediction of each row using the "classify_one" function
    df["classification"] = df.apply(classify_one, axis=1, args=(tree,))
    
    #If classification for row is correct, add it to df["classification_correct"]
    df["classification_correct"] = df["classification"] == df["income"]
    
    # Calculate the accuracy using mean of total predictions
    accuracy = df["classification_correct"].mean()
    
    return accuracy
```

## 1. decision tree classifier based on infogain

```python
#parameters
#df: the dataframe containing the data to be trained and classified
#max_depth: the maximum depth of the decision tree/the maximum number of splits
#min_samples: the minimum number of samples required for a split to occur
#counter: a counter variable to keep track of the depth of the tree
def decision_tree_classifier_infogain(df, max_depth, min_samples, counter=0):
    
    #initialize variables at the start
    if counter == 0:
        global COLUMN_HEADERS
        COLUMN_HEADERS = df.columns
        data = df.values
    else:
        data = df           
    
    # early stopping conditions
    # if the data is a leaf node/min samples > number of samples in data/max depth reached = get leaf node classification
    if (check_leaf(data)) or (len(data) < min_samples) or (counter == max_depth):
        classification = classify_data(data)
        
        return classification
    
    #Else, get all possible splits and choose the best split with best info gain
    #Then it splits the data into two branches, left and right, based on the best split and the question generated from it.
    else:    
        # increment the counter for each iteration
        counter += 1
        # get all possible splits
        possible_splits = get_possible_splits(data)
        
        #returns the feature/col with best info gain
        split_column, split_value = determine_best_split_infogain(data, possible_splits)
        
        #split data into left and right branches using best feature as decided by info gain
        left, right = split_data(data, split_column, split_value)
        
        # choose question (key)
        # get the name of the feature that was used for the best split
        feature_name = COLUMN_HEADERS[split_column]
            
        #format question (key) (e.g. "feature_name = split_value")
        question = "{} = {}".format(feature_name, split_value)
        
        #initialise dictionary for storing tree
        sub_tree = {question: []}
    
        #recursive call with left and right branches
        yes_answer = decision_tree_classifier_infogain(left, max_depth , min_samples, counter)
        no_answer = decision_tree_classifier_infogain(right, max_depth, min_samples, counter)
        
        #if left and right branches values is same, assign ans to dictionary
        if yes_answer == no_answer:
            sub_tree = yes_answer
        #else return the sub_tree dictionary.
        else:
            sub_tree[question].append(yes_answer)
            sub_tree[question].append(no_answer)
            
    #return final decision tree as a dictionary
    return sub_tree
```

## 4. Ensemble method to build new classifier

```python
#Utilises voting function on prediction outcome of the 3 dt models before.

#Parameters
#prediction: an array containing the predictions of an individual model (info gain)
#prediction1: an array containing the predictions of an individual model (gini index)
#prediction2: an array containing the predictions of an individual model (gain ratio)
def ensemble_voting(prediction, prediction1, prediction2):

    #Add all models and predictions to dataframe
    all_predictions = pd.DataFrame({'model1': prediction,
                           'model2': prediction1,
                           'model3': prediction2})
    
    #Find highest occurance per column (Voting function)
    ensemble_predictions = all_predictions.mode(axis=1)
    
    #returns a numpy array containing the ensemble predictions
    return np.array(ensemble_predictions[0])
```
